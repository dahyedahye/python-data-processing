```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(1, figsize = (10,10))
x = np.arange(0.0,1.0,0.0001)
x = np.linspace(-4.5,4.5,num=10000)
y = np.log(1-x)

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, linewidth=2.5)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.xlim(-4.1, 1.1)
plt.ylim(-3.1, 2.1)

ax = plt.gca()
ax.set_aspect(1)
# plt.savefig('log_1_d.pdf', bbox_inches='tight' ) # save the graph image in pdf format

plt.show()
```

```python
fig = plt.figure(1, figsize = (10,10))
x = np.arange(-2.0,2.0,0.0001)
x = np.linspace(-5.5,5.5,num=10000)

y = -(np.log(x))

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, linewidth=2.5)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.tick_params(axis="x", labelsize=15)
ax.tick_params(axis="y", labelsize=15)
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.xlim(-0.1, 5.1)
plt.ylim(-2.1, 3.1)

ax = plt.gca()
ax.set_aspect(1)

# plt.savefig('minus_log_d.pdf', bbox_inches='tight' ) # save the graph image in pdf format

plt.show()
```
