```python
import numpy as np
import pandas as pd
import glob
import re


dir_folder = './sample_data' # dir of a folder that contains txt files
file_list = [f for f in glob.glob("{}/*.txt".format(dir_folder))] # append all files with .txt extension

train_configs = ['balanced']
models = ['efficientnet-b3', 'resnet18', 'resnet152', 'resnext101']
scores = ['test_acc']
dict_result_all = {}
dict_result_hair = {}

for tc in train_configs:
    for m in models:
        for s in scores:
            list_acc_tmp_all = []
            list_acc_tmp_hair = []
            for fl in file_list:
                if (tc in fl) and (m in fl) and (s in fl):
                    file_txt = open(fl, 'r')
                    lines = file_txt.readlines() # read all lines in the .txt file line by line
                    for l in lines:
                        if 'acc' in l and 'hair acc' not in l:
                            p = re.compile(r'\([^)]*\)')
                            l = re.sub(p, '', l) # remove unnecessary part in parenthesis like ([epoch]584)
                            acc_value = l.replace("acc : ", "") # remove unnecessary part "acc :"
                            acc_value = acc_value.replace(" ", "") # remove unnecessary space
                            acc_value = acc_value.replace("\n", "") # remove unnecessary space
                            acc_value = float(acc_value) # convert datatype to float
                            list_acc_tmp_all.append(acc_value)
                            list_acc_tmp_all = sorted(list_acc_tmp_all, key = lambda x:float(x)) 
                        elif 'hair acc' in l:
                            p = re.compile(r'\([^)]*\)')
                            l = re.sub(p, '', l) # remove unnecessary part in parenthesis like ([epoch]584)
                            acc_value = l.replace("hair acc : ", "") # remove unnecessary part "acc :"
                            acc_value = acc_value.replace(" ", "") # remove unnecessary space
                            acc_value = acc_value.replace("\n", "") # remove unnecessary space
                            acc_value = float(acc_value) # convert datatype to float
                            list_acc_tmp_hair.append(acc_value)
                            list_acc_tmp_hair = sorted(list_acc_tmp_hair, key = lambda x:float(x)) 
                            
            key_name = '{}_{}_{}'.format(tc, m, s)
            dict_result_all[key_name] = list_acc_tmp_all
            dict_result_hair[key_name] = list_acc_tmp_hair

# show all columns in df
pd.options.display.max_columns = 0
df_all = pd.DataFrame.from_dict(dict_result_all)
df_all.describe()
```
